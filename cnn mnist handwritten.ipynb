```python
# convolutional neural network on MNIST dataset
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('mnist.csv')
```

```python
data.head()
```

```python
#reshaping into 28x28 because its  black and white images
data.iloc[3,1:].values.reshape(28,28).astype('uint8')
```

```python
#processing data
```

```python
#storing pixel array in form length 28, width 28, and channel 1 in df_x
df_x = data.iloc[:, 1:].values.reshape(len(data),28,28,1)

#storing the lebels in y
y= data.iloc[:,0].values
```

### convert labels into catogorical data 
### convert labels to catagorical and each colum represents it own values declared above i.e 10

```python
import keras
```

```python
df_y = keras.utils.to_categorical(y,num_classes = 10)
```

```python
df_x = np.array(df_x)
df_y - np.array(df_y)
```

```python
y
```

```python
df_y
```

```python
df_x.shape
```

```python
#test and train

x_train, x_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.2, random_state=4)
```

## CNN

```python
model = Sequential()
model.add(Convolution2D(32,3,data_format='channels_last', activation = 'relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss= 'categorical_crossentropy', optimizer='adadelta',metrics=['accuracy'])
```

```python
model.summary()
```

```python
#fittingit with 100 images for testing
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs = 10)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.predict(x_test)
```
